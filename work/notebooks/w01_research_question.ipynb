An SEO editor acts on this output by editing whichever page the ranking points them to, rather than relying only on FlyRank's existing rule-based flags.  
A wrong call costs in two directions:
- if the model sends the editor to a page that didn't need fixing, that's roughly 2-3 hours of wasted editor time per false alarm (across ~10 pages/week, that adds up fast)
- if the model misses a page that's genuinely declining, the cost is lost traffic and clicks that keep bleeding until someone notices.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

total_pages = len(df)
declining = df[df["trend_direction"] == "down"]
n_declining = len(declining)
pct_declining = 100 * n_declining / total_pages

clicks_declining = declining["clicks_last_30d"].sum()
clicks_total = df["clicks_last_30d"].sum()
impressions_declining = declining["impressions_last_30d"].sum()

print(f"Total pages: {total_pages:,}")
print(f"Declining pages (trend_direction == 'down'): {n_declining:,} ({pct_declining:.1f}%)")
print(f"Clicks (last 30d) still landing on declining pages: {clicks_declining:,} of {clicks_total:,} total ({100*clicks_declining/clicks_total:.1f}%)")
print(f"Impressions (last 30d) on declining pages: {impressions_declining:,}")

# Why ranking matters, not just "fix everything declining":
editor_pages_per_week = 10
weeks_to_clear_backlog = n_declining / editor_pages_per_week
print(f"\nAt ~{editor_pages_per_week} pages/week, clearing the declining backlog unordered would take ~{weeks_to_clear_backlog:.0f} weeks (~{weeks_to_clear_backlog/52:.1f} years).")
print("That's the case for ranking: with a backlog this size, WHICH page first is the real decision.")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

### What this work will be able to claim:
- An observed, decision-support ranking: "these pages show measured decline in this 90-day window, and rank here relative to others."
- A tool that helps an SEO editor prioritize - not a verdict, a suggestion.
###What this work will NEVER claim:
- That it predicts what Google's ranking algorithm will do.
- That a high-ranked page is guaranteed to decline further, or that fixing it is guaranteed to help - these are observational signals, not causal proof.
